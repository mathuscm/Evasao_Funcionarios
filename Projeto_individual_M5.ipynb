# 4. Crie três maneiras diferentes de mostrar esses dados e liste quais são algumas vantagens e insuficiências de cada uma.


## Gráfico de barras horizontais



horizontal2.jpghorizontal1.jpg

Vantagens:
- Boa visualização dos rótulos e valores em uma quantidade compacta de dados,
- boa noção da diferença entre os dados tanto pelas barras quanto pelo eixo horizontal,
- largura uniforme das barras,
- eficaz para exibir comparação entre categorias.

Desvantagens:
- Se não estiver discriminado que o eixo y está em ordem de impacto (do maior 'training' para o menor 'illness') pode prejudicar o manejamento do gráfico, assim como a interpretação e tomada de decisão,
- se houver muitas categorias a leitura dos rótulos fica prejudicada deixando o texto comprimido e ilegível.

## Gráfico de barras vertical tríplice

vertical-triplice.jpg

Vantagens:
- Comparação direta dos elementos qualitativo (nível de impacto dos tópicos) e quantitativo (pesquisa do RH e formulário de saída),
- ênfase à comparação já que ao agrupar os três tipos de dados e diferenciá-los por cor, as discrepancias se tornam mais claras,

- Aproveitamento de espaço, considerando o agrupamento em um só quadrante, não se torna necessário outro gráfico apenas para dois elementos, tornando viualmente mais prático

- Visualização rápida do comportamento dos dados

Desvantagens:
- Comparação limitada de dados para evitar excessos e não tornar a visualização complexa seja pela quantidade de categorias/barras e distância entre elas,

- É necessário utilizar cores distintas para identificar os diferentes tipos de dados, reiterando a quantidade amigável de dados a serem comparados afim de evitar poluição, pois seria prejudicial para a discriminação e interpretação
